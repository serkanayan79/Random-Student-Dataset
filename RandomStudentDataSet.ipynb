```python
import numpy as np
import pandas as pd

# Öğrenci sayısı
num_students = 100

# Okul yılları
school_years = np.random.choice([1, 2, 3, 4], size=num_students)

# Okul bağlantıları
affiliations = np.random.choice([
    "Boğaziçi University",
    "Gazi University",
    "ODTU",
    "ITU",
    "Hacettepe University",
    "Gebze Technical University",
    "İzmir Yüksek Teknoloji Üniversitesi",
    "Harran Üniversitesi",
    "Balikesir Üniversitesi",
    "Bursa Teknik Üniversitesi",
    "Uludağ Üniversitesi",
    "Bilkent Üniversitesi",
    "Koç Üniversitesi",
    "Sabanci Üniversitesi"
], size=num_students)

# İlgi alanları
interests = np.random.choice([
    "Quantum Theory",
    "Complex Systems/Dynamical Systems",
    "Analysis",
    "Topology",
    "Algebra",
    "Partial Differential Equations",
    "Cryptography/Coding theory",
    "Probability Theory",
    "Complex Analysis",
    "Real Analysis/Functional Analysis",
    "Algebraic Number Theory",
    "Classical Mechanics",
    "Fluid Dynamics",
    "Hydrodynamic Instability",
    "Mathematical Modeling",
    "Data Science",
    "Statistics"
], size=num_students)

# Cinsiyet
genders = np.random.choice(["Kadin", "Erkek"], size=num_students)


data = pd.DataFrame({
    "School year": school_years,
    "Affiliation": affiliations,
    "Branches of Interest": interests,
    "Cinsiyet": genders
})

# Veriyi CSV dosyası olarak kaydet
data.to_csv("ogrenci_veri_seti2.csv", index=False)
```
